```python
import skimage as si
import numpy as np
from skimage.io import imread, imshow,imsave
import matplotlib.pyplot as plt
import random
import os
```

```python
path = "F:\Major Project\Multiple-Encryption-Algorithms-And-LSB-Steganography\Auxillary\Split File\Images\demo.jpg"
```

```python
def splitImage(path,sections= random.randint(3,6),save = False, dest_path = os.getcwd()):
    
    image  = imread(fname=path)
    lis = np.split(image,sections)
    
    if(save):
        for i in range(sections):
            imsave(dest_path+"\\part {0}.jpg".format(i),lis[i])
            
    return lis        
    
    
    
    
    
```

```python
def mergeImage(lis,show = False):
    image = np.vstack(lis)
    if(show):
        imshow(image)
    
    return image
```

```python
ls = splitImage(path = path,save = True,dest_path ="C:\\Users\\hp\\Desktop")
```

```python
mergeImage(ls,True)
```
